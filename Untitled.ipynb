```python
#Import libraries
import pandas as pd
import seaborn as sns

#Load and read df
stocktweets = pd.read_csv("stocktweet.csv")
stocktweets
```

```python
# Check and drop null values
stocktweets.isnull().sum()
```

```python
# Remove duplicates
stocktweets.drop_duplicates(inplace=True)
stocktweets
```

```python
import os

stockprice_folder = 'stockprice'
stockprice_dfs = {}

for filename in os.listdir(stockprice_folder):
    if filename.endswith('.csv'):
        data = filename.split('.')[0]
        df = pd.read_csv(os.path.join(stockprice_folder, filename))
        stockprice_dfs[data] = df
```

```python
for data, df in stockprice_dfs.items():
    # Check for missing values
    print(data, df.isnull().sum())

    # Handle missing values if necessary (e.g., fill with mean)
    stockprice_dfs[data].fillna(method='ffill', inplace=True)
```

```python
import re
# Clean tweets by removing unprintable characters using regex
check = r'[^a-zA-Z0-9]'
stocktweets['tweet'] =stocktweets['tweet'].str.replace(check, ' ', regex=True)
```
